```python
# Saee Surve
# Experiment: Load Balancing Simulation - Round Robin & Least Connections

import random
import time

# ─── Server Setup ─────────────────────────────────────────────────────────────
class Server:
    def __init__(self, name):
        self.name            = name
        self.active_conn     = 0
        self.total_handled   = 0

    def handle_request(self, req_id):
        self.active_conn   += 1
        self.total_handled += 1
        print(f"  → Request {req_id:02d} assigned to {self.name} | Active: {self.active_conn} | Total: {self.total_handled}")

    def release(self):
        if self.active_conn > 0:
            self.active_conn -= 1

    def reset(self):
        self.active_conn   = 0
        self.total_handled = 0

    def status(self):
        return f"{self.name}: total={self.total_handled}, active={self.active_conn}"


# ─── Round Robin ──────────────────────────────────────────────────────────────
class RoundRobin:
    def __init__(self, servers):
        self.servers = servers
        self.index   = 0

    def get_server(self):
        server      = self.servers[self.index]
        self.index  = (self.index + 1) % len(self.servers)
        return server

    def distribute(self, num_requests):
        print("\n" + "="*50)
        print("       ROUND ROBIN LOAD BALANCING")
        print("="*50)
        for req_id in range(1, num_requests + 1):
            server = self.get_server()
            server.handle_request(req_id)
            # simulate random request completion
            if random.random() > 0.5:
                server.release()

        print("\n--- Final Server Status ---")
        for s in self.servers:
            print(" ", s.status())


# ─── Least Connections ────────────────────────────────────────────────────────
class LeastConnections:
    def __init__(self, servers):
        self.servers = servers

    def get_server(self):
        return min(self.servers, key=lambda s: s.active_conn)

    def distribute(self, num_requests):
        print("\n" + "="*50)
        print("     LEAST CONNECTIONS LOAD BALANCING")
        print("="*50)
        for req_id in range(1, num_requests + 1):
            server = self.get_server()
            server.handle_request(req_id)
            # simulate random request completion
            if random.random() > 0.5:
                server.release()

        print("\n--- Final Server Status ---")
        for s in self.servers:
            print(" ", s.status())


# ─── Main ─────────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    NUM_REQUESTS = 10

    # Round Robin
    servers_rr = [Server("Server-A"), Server("Server-B"), Server("Server-C")]
    rr = RoundRobin(servers_rr)
    rr.distribute(NUM_REQUESTS)

    # Least Connections
    servers_lc = [Server("Server-A"), Server("Server-B"), Server("Server-C")]
    lc = LeastConnections(servers_lc)
    lc.distribute(NUM_REQUESTS)

    # ─── Comparison Summary ───────────────────────────────────────────────────
    print("\n" + "="*50)
    print("           COMPARISON SUMMARY")
    print("="*50)
    print(f"{'Server':<12} {'RR Total':>10} {'LC Total':>10}")
    print("-"*35)
    for s_rr, s_lc in zip(servers_rr, servers_lc):
        print(f"{s_rr.name:<12} {s_rr.total_handled:>10} {s_lc.total_handled:>10}")
```
